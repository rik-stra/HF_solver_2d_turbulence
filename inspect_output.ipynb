```python
import matplotlib.pyplot as plt
import numpy as np
import h5py
```

```python
fname = "./output/samples/From_initial_conditions_t_10.0.hdf5"
```

```python
with h5py.File(fname, "r") as f:
    w_hat = f["w_hat_n_HF"] #fourier transform of the vorticity solution [t, N, N]
    w = np.fft.ifft2(w_hat).real #vorticity fields [t, N, N]
```

```python
# contour plot (create a plot based on isocontours of the vorticity field, might be smoother than the true field)
cb = plt.contourf(w[10,:,:], 100)
plt.colorbar(cb)
```

```python
# image plot   (this shows the value per array index)
plt.imshow(w[10,:,:], cmap='viridis')
```
